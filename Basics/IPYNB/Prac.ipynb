```python
class student:
    def list(self):
        self.students =[]

    def add(self, id, name, age):
        students = {
            'id':id,
            'name':name,
            'age':age
        }
        self.students.append(students)
    
    def display(self):
        for student in self.students:
            print(f"id: {student['id']}")
            print(f"name: {student['name']}")
            print(f"age: {student['age']}")
            print()

    def update(self, id, name, age):
        for student in self.students:
            if student['id']==id :
                student['name'] = name
                student['age'] = age
                print(f"Student with Roll Number {id} updated.")
                return
        print(f"Student with Roll Number {id} not found.")

    def search(self, id):
        for student in self.students:
            if student['id'] == id:
                print("Student Found:")
                print(f"id: {student['id']}")
                print(f"name: {student['name']}")
                print(f"age: {student['age']}")
                return
        print(f"Student with Roll Number {id} not found.")

a = student()

while True :
    print("Student Details Menu:")
    print("1. Add Student")
    print("2. Display Students")
    print("3. Update Student")
    print("4. Search Student")
    print("5. Quit")

    choice = int(input("enter the choice you want"))

    if choice == 1:
        id = input("Enter Roll Number: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        a.add(id, name, age)

    elif choice == 2:
        a.display()

    elif choice == 3:
        id = input("Enter Roll Number: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        a.update(id,name,age)

    elif choice == 4:
        id = input("Enter the roll number")
        a.search(id)
    
    elif choice == 5:
        break

    else:
        print("Invalid Choice")


```

```python
class StudentDetails:
    def __init__(self):
        self.students = []

    def add_student(self, roll_number, name, age):
        student = {
            'roll_number': roll_number,
            'name': name,
            'age': age
        }
        self.students.append(student)

    def display_students(self):
        if not self.students:
            print("No student records found.")
            return

        print("Student Records:")
        for student in self.students:
            print("Roll Number :",student['roll_number'])
            print("Name:",student['name'])
            print("Age:",student['age'])
            print()

    def update_student(self, roll_number, name, age):
        for student in self.students:
            if student['roll_number'] == roll_number:
                student['name'] = name
                student['age'] = age
                print("Student with Roll Number" + roll_number + "updated.")
                return
        print("Student with Roll Number" +  roll_number + "not found.")

    def search_student(self, roll_number):
        for student in self.students:
            if student['roll_number'] == roll_number:
                print("Student Found:")
                print("Roll Number:" + student['roll_number'])
                print("Name:" + student['name'])
                print("Age:" + student['age'])
                return
        print("Student with Roll Number" + roll_number + "not found.")

# Menu-driven program
student_data = StudentDetails()

while True:
    print("Student Details Menu:")
    print("1. Add Student")
    print("2. Display Students")
    print("3. Update Student")
    print("4. Search Student")
    print("5. Quit")

    choice = input("Enter your choice: ")

    if choice == '1':
        roll_number = input("Enter Roll Number: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        student_data.add_student(roll_number, name, age)

    elif choice == '2':
        student_data.display_students()

    elif choice == '3':
        roll_number = input("Enter Roll Number to update: ")
        name = input("Enter Updated Name: ")
        age = input("Enter Updated Age: ")
        student_data.update_student(roll_number, name, age)

    elif choice == '4':
        roll_number = input("Enter Roll Number to search: ")
        student_data.search_student(roll_number)

    elif choice == '5':
        print("Exiting the program.")
        break

    else:
        print("Invalid choice. Please enter a valid option.")
```
