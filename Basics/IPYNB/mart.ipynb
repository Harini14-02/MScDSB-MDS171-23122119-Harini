```python
class bakemart:
    def __init__(self):
        self.inventory = {}
        self.order = {}

    def create_order(self, id, product, qty, price, total):
        temp = {
            "Id": id,
            "Product": product,
            "Qty": qty,  # Convert to int
            "Price": price,  # Convert to float
            "total": total,  # Convert to float
        }
        self.order[id] = temp

    def create_inventory(self, id, product, price):
        temp1 = {
            "Id": id,
            "Product": product,
            "Price": price,  # Convert to float
        }
        self.inventory[id] = temp1

    def print_orders(self):
            print(self.inventory)

    def print_inventory(self):
            print(self.order)

    def new_order(self):
        id = str(input("Enter the order ID: "))
        product = input("Enter the product name: ")
        qty = input("Enter the quantity: ")
        price = input("Enter the price: ")
        total = input("Enter the total: ")
        self.create_order(id, product, qty, price, total)

trinity = bakemart()
orders = open("Book1.csv", "r")
orders1 = open("Inventory1.csv", "r")
orders_header = orders.readline()
orders_order = orders.readlines()
orders1_header = orders1.readline()
orders1_order = orders1.readlines()

for item in orders_order:
    temp = item.strip().split(',')
    trinity.create_order(temp[0], temp[1], temp[2], temp[3], temp[4])

trinity.print_orders()
trinity.print_inventory()

for item in orders1_order:
    temp1 = item.strip().split(',')
    trinity.create_inventory(temp1[0], temp1[1], temp1[2])

def new_order():
    id = input("Enter the order ID: ")
    product = input("Enter the product name: ").title()
    qty = input("Enter the quantity: ")
    price = input("Enter the price: ")
    total = input("Enter the total: ")
    inv = open("Inventory1.csv", 'a+')
    for item_id, item_info in trinity.inventory.items():
        if item_info['Product'] == product:
            current_qty = item_info['Qty']
            new_qty = current_qty - int(qty)
            item_info['Qty'] = new_qty
            inv.write(f"{item_id},{item_info['Product']},{new_qty},{item_info['Price']}\n")

    file = open("Book1.csv", "a+")
    file.write(f"{id},{product},{qty},{price},{total}\n")
    file.close()
    trinity.create_order(id, product, qty, price, total)

new_order()

```
